```python
class Aircraft:
    def __init__(self, weight):
        self._weight = weight

class UAV:
    def __init__(self):
        self._has_autopilot = True
        self._missions = []  # список миссий

    #добавлен код для декоратора атрибута _missions
    @property
    def missions(self):
        return self._missions

    @missions.setter
    def missions(self, mission):
        self._missions.append(mission)

    # добавлен публичный метод count_missions, который возвращает количество миссий
    def count_missions(self):
        return len(self._missions)

class MultirotorUAV(Aircraft, UAV):
    def __init__(self, weight, model, brand):
        super().__init__(weight)
        UAV.__init__(self)
        self._model = model
        self._brand = brand

    # добавлен публичный метод get_info
    def get_info(self):
        return f"Model: {self._model}, Brand: {self._brand}, Weight: {self._weight}"

    # добавлен публичный метод get_model
    def get_model(self):
        return self._model
```

**TODO 3-1** - Добавить атрибут incidents типа список

Добавьте в класс MultirotorUAV атрибут incidents и внесите туда информацию обо всех найденных происшествиях для этой модели. 
Не забудьте, что атрибут добавляется при помощи декоратора


```python
class Aircraft:
    def __init__(self, weight):
        self._weight = weight

class UAV:
    def __init__(self):
        self._has_autopilot = True
        self._missions = []

    # декоратор возвращает список миссий
    @property
    def missions(self):
        return self._missions 

    # добавляет миссию в список миссий
    @missions.setter
    def missions(self, mission):
        self._missions.append(mission)

    # количество миссий 
    def count_missions(self):
        count = len(self.missions)
        return count        

class MultirotorUAV(Aircraft, UAV):

    def __init__(self, weight, model, brand):
        super().__init__(weight)
        UAV.__init__(self)
        self.__weight = weight
        self.__brand = brand
        self.__model = model
      
        # добавьте приватный атрибут incidents
        self.__incidents = []

    def get_info(self):
        return f"масса: {self.__weight}, производитель: {self.__brand}, количество миссий: {self.count_missions()}"

    def get_model(self):
        return self.__model
      
    def get_weight(self):
       return self.__weight

    def get_brand(self):
      return self.__brand

	# добавлен код декоратора для атрибута incidents
    @property
    def incidents(self):
        return self.__incidents

	# добавлен публичный метод add_incident, который добавляет инцидент в список инцидентов для данной модели дрона
    def add_incident(self, incident):
        self.__incidents.append(incident)

	# добавлен публичный метод save_data, который сохраняет информацию о дроне в файл json
    def save_data(self):
        path = self.__class__.__name__ + "_" + self.get_model() + ".json"
        data = {
            "weight": self.get_weight(),
            "model": self.get_model(),
            "brand": self.get_brand(),
            "incidents": self.incidents,
            "missions": self.missions
        }
        with open(path, 'w') as file:
            json.dump(data, file, indent=4)

```

Прочий необходимый код (**взят из TODO 3 Работы №2**)

```python
def read_json(filename): # чтение json
    with open(filename) as json_file:
        data = json.load(json_file)
    return data

def missions_by_pilot(pilot, data): # функция возвращает список миссий для указанного пилота pilot из данных data.
    return data[pilot]["missions"]

def missions_by_pilots(data): # функция возвращает словарь, где ключи - пилоты, значения - количество миссий
    missions_count_by_pilots = {}
    for pilot in data.keys():
        missions = missions_by_pilot(pilot, data)
        missions_count_by_pilots[pilot] = len(missions)
    return dict(sorted(missions_count_by_pilots.items(), key=lambda item: item[1], reverse=True))

def missions_by_drones(data): # функция возвращает словарь, где ключи -  модели дронов, значения - списки миссий
    drones_missions_dict = {}
    for pilot in data.keys():
        missions = missions_by_pilot(pilot, data)
        for mission in missions:
            drone = mission["drone"]
            if not drone in drones_missions_dict.keys():
                drones_missions_dict[drone] = []    
            drones_missions_dict[drone].append(mission["mission"])
    return drones_missions_dict 

def types(data): # функция возвращает список уникальных моделей дронов
    models = []
    drones_missions = missions_by_drones(data)
    for drone in drones_missions.keys():
        if not drone in models:
            models.append(drone)
    return models           

def count_missions(data): # функция возвращает словарь, где ключи -  модели дронов, значения - количество миссий
    drones_missions = missions_by_drones(data)
    result = {}
    for drone in drones_missions.keys():
        result[drone] = len(drones_missions[drone])
    return result    
```

```python
drone_catalog = {
	"DJI Mavic 2 Pro": {"weight":903, "brand":"DJI"},
	"DJI Mavic 2 Zoom": {"weight":900, "brand":"DJI"},
	"DJI Mavic 2 Enterprise Advanced": {"weight":920, "brand":"DJI"},
	"DJI Inspire 2": {"weight":1500, "brand":"DJI"},
	"DJI Mavic 3": {"weight":1000, "brand":"DJI"}
}
```
